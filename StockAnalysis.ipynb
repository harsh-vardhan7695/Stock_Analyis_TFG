## Moving Avg of GM

```python
gm['Open'].plot(label='GM Open',figsize=(16,7))
```

#### Here the data looks noisy,so data might not be reliable given short observation window.SO here Moving Avg comes into picture.Its a smoothing method in time series,its a way to make your data more reliable and robust against noise

```python
gm['Open'].plot(label='No Moving AVG',figsize=(16,7))
gm['MAvg50']=gm['Open'].rolling(50).mean()
gm['MAvg50'].plot(label='Moving AVG 50')
gm['MAvg100']=gm['Open'].rolling(100).mean()
gm['MAvg100'].plot(label='Moving AVG 100')
plt.legend(loc='upper left')
```

#### When moving avg window size is high,you might miss out some trends

### Daily Percentage change

```python
tesla['returns']=(tesla['Close']/tesla['Close'].shift(1))-1
```

```python
ford['returns']=(ford['Close']/ford['Close'].shift(1))-1
```

```python
ford.head()
```

```python
ford['returns'].hist(bins=50)
```

#### this looks like a bell curve,that is normally distributed.Its not so much volatile 

```python
gm['returns'].hist(bins=50)
```

```python
tesla['returns'].hist(bins=50)
```

```python
plt.figure(figsize=(15,6))
tesla['returns'].hist(bins=50,label='Tesla',alpha=0.5)
gm['returns'].hist(bins=50,label='GM',alpha=0.5)
ford['returns'].hist(bins=50,label='FORD',alpha=0.5)

plt.tight_layout
plt.legend()
plt.show
```

#### Here we can see that curve cooresponding to TESLA is wider as compared to GM and FORD ,that beacuse TESLA stocks are more volatile in nature.GM and FORD are very similar in terms of volatility and normally distributed

```python
# plt.figure(figsize=(15,6))
box=pd.concat([tesla['returns'],gm['returns'],ford['returns']],axis=1)
box.columns=['Tesla Returns','GM Returns','Ford Returns']
box.plot(kind='box',figsize=(15,6))
# plt.legend()
plt.show()
```

#### As you can see the daily percentage of tesla is spread out...and OUTLIERS are also present.But GM and Ford are normally distributed over this period of time

### Lets see again the correlation b/w the these stocks in respect to Daily percentage retuns,we will use Scatter PLot matrix as done above

```python
pd.plotting.scatter_matrix(box,figsize=(8,8),hist_kwds={'bins':50},alpha=0.25)
```

#### For GM and Ford we can see a linear type of cluster,so they do have possible relationship.And when we see Tesla and GM we can see a circular cluster,i.e no Corelation

## Cumulative Return

#### numpy.cumprod() function is used when we want to compute the cumulative product of array elements over a given axis.

```python
gm['cr']=(1+gm['returns']).cumprod()
```

```python
gm.head()
```

```python
ford['cr']=(1+ford['returns']).cumprod()
ford.head()
```

```python
tesla['cr']=(1+tesla['returns']).cumprod()
tesla.head()
```

```python
plt.figure(figsize=(16,7))
tesla['cr'].plot(label='Tesla')
plt.legend(loc='upper left')
```

```python
plt.figure(figsize=(16,7))
gm['cr'].plot(label='GM')
plt.legend(loc='upper left')
```

```python
plt.figure(figsize=(16,7))
ford['cr'].plot(label='FORD')
plt.legend(loc='upper left')
```

```python
plt.figure(figsize=(16,7))
tesla['cr'].plot(label='TESLA')

ford['cr'].plot(label='FORD')
gm['cr'].plot(label='GM')
plt.title('Cummualtive Return')
plt.xlabel('YEAR')
plt.ylabel('How muchh times ')
plt.legend(loc='upper left')
```

#### This clearly show that if we would have invested 1$ in 2012 in all the stocks...the maximum return would come from TESLA
